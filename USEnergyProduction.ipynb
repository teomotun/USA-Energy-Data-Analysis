```python
response = requests.get(url).json()
```

```python
years = []
values = []

for x in response["series"][0]["data"]:
    years.append(x[0])
    values.append(x[1])
    
#values = response["series"][0]["data"][1]
years  = years[::-1]
values = values[::-1]
```

```python
plt.figure(figsize=(20,10))
plt.scatter(years, values)


#plt.grid(True)

plt.show()
```
